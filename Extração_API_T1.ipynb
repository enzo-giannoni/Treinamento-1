```python
import pandas as pd
import numpy as np
import requests
```

```python
def acessa_api(code):
    headers = {'user-agent': 'Mozilla/5.0'}
    resposta = requests.get(f'https://treinamento-ndados.s3.sa-east-1.amazonaws.com/boardgames_{code}_raw.json', headers = headers)

    try:
        assert resposta.status_code == 200

    except AssertionError: 
        return []

    return resposta.json()

info = acessa_api('information')
categ = acessa_api('categoricals')
stat = acessa_api('stat')
```

## Base de Informações

| Nome da Feature   | Descrição      |
|-------------------|----------------|
| objectid          | [int] id do game em boardgamegeeks.com, usado para linkar tabelas   |
| name              | [String] O nome do game               |
| yearpublished     | [int] O ano de publicação do game  |

| minplaytime       | [int] mínimo de tempo de jogo |
